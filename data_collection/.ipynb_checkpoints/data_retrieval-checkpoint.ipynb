```python
import pandas as pd
import collections
from eventregistry import *
```

```python
er = EventRegistry()
all_sources = ["The New York Times",
               "Washington Post",
               "San Francisco Gate", 
               "Breitbart",
               "Fox News",
               "National Review",
               "The American Conservative"]
q = QueryArticlesIter(conceptUri = er.getConceptUri("Politics"), lang = "eng", sourceUri = QueryItems.OR(
    [er.getNewsSourceUri(source) for source in all_sources]))
```

```python
res = q.execQuery(er, sortBy = "date")

print("Number of articles:", q.count(er))
```

```python
titles = []
sources = []
contents = []
for art in res:
    titles.append(art['title'])
    sources.append(art['source']['title'])
    contents.append(art['body'])
```

```python
print(collections.Counter(sources))

df = pd.DataFrame(
    {'title': titles,
     'source': sources,
     'body': contents,
     'index': range(len(titles))
    })

df.to_csv('articles.csv', index=False)
```
